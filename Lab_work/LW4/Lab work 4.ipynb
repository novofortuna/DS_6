2. Отримати курс долара за останній рік.

```python
from datetime import datetime, timedelta
from statistics import mean, stdev
import requests
import pickle 

a = datetime.today()
d = datetime(int(a.year)-1, a.month, a.day)
Course = []

for i in range((a-d).days):
    p = {'date' : d.strftime('%Y%m%d'), 'valcode' : 'USD'}
    r = requests.get('https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?json', params = p)
    if r.status_code ==200:
        Course.append(r.json()[0]['rate'])
        d = d + timedelta(days=1)
print(f'Курс USD за останній рік: \n{Course}')
with open('Course_USD.text', 'w') as f:
    f.write(f'Курс USD за останній рік: \n{Course}\n')
```

3. Вивести середнє значення та відхилення курсу за кожний місяць.

```python
from datetime import datetime, timedelta
from statistics import mean, stdev
import requests
import pickle

a = datetime.today()
d = datetime(int(a.year)-1, a.month, 1)
DK = []

for i in range(12):
    if d.month < 12:
        d1 = datetime(d.year, int(d.month)+1, 1)
        day = d1-d
        for i in range(day.days):
            p = {'date' : d.strftime('%Y%m%d'), 'valcode' : 'USD'}
            r = requests.get('https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?json', params = p)
            if r.status_code ==200:
                DK.append(r.json()[0]['rate'])
                d = d + timedelta(days=1)
        print(f'Щоденний курс USD за {d.month-1} місяць {d.year} року: \n{DK}')
        print(f'Середнє значення USD за {d.month-1} місяць {d.year} року становить = {round(mean(DK), 4)}')
        print(f'Відхилення курсу USD за {d.month-1} місяць {d.year} року становить = {round(stdev(DK), 4)}\n')
        with open('Course_USD.text', 'a') as f:
            f.write(f'Щоденний курс USD за {d.month-1} місяць {d.year} року: \n{DK}\n')
            f.write(f'Середнє значення USD за {d.month-1} місяць {d.year} року становить = {round(mean(DK), 4)}\n')
            f.write(f'Відхилення курсу USD за {d.month-1} місяць {d.year} року становить = {round(stdev(DK), 4)}\n')
        DK = []
        if d1.month == a.month:
            break

    if d.month == 12:
        d1 = datetime(int(d.year)+1, 1, 1)
        day = d1-d
        for i in range(day.days):
            p = {'date' : d.strftime('%Y%m%d'), 'valcode' : 'USD'}
            r = requests.get('https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?json', params = p)
            if r.status_code ==200:
                DK.append(r.json()[0]['rate'])
                d = d + timedelta(days=1)
        print(f'Щоденний курс USD за 12 місяць {d.year-1} року: \n{DK}')
        print(f'Середнє значення USD за 12 місяць {d.year-1} року становить = {round(mean(DK), 4)}')
        print(f'Відхилення курсу USD за 12 місяць {d.year-1} року становить = {round(stdev(DK), 4)}\n')
        with open('Course_USD.text', 'a') as f:
            f.write(f'Щоденний курс USD за 12 місяць {d.year-1} року: \n{DK}\n')
            f.write(f'Середнє значення USD за 12 місяць {d.year-1} року становить = {round(mean(DK), 4)}\n')
            f.write(f'Відхилення курсу USD за 12 місяць {d.year-1} року становить = {round(stdev(DK), 4)}\n')
        DK = []
```

### Завдання 2:

Потрібно проаналізувати всі товари на сайті: 
https://smallpacking.agrosem.ua/shop/

Створіть базу даних, що містить таблицю із такими стовбцями: 
    - артикул товару
    - найменування
    - ціна 
    - вага

```python
HEADERS = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:49.0) Gecko/20100101 Firefox/49.0',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
            'Accept-Language': 'en-US,en;q=0.5',
            'Accept-Encoding': 'gzip, deflate',
            'DNT': '1',
            'Connection': 'keep-alive',
            'Upgrade-Insecure-Requests': '1'
        }

import requests
from bs4 import BeautifulSoup

BASE = []

SITE = 'https://smallpacking.agrosem.ua/shop/'

def pars1():
    r = requests.get(SITE, headers = HEADERS)

    soup = BeautifulSoup(r.text, 'lxml')

    for el in soup.find('div', class_ = 'product-category-wrapper').find_all('div', class_='product-from-category-container'):
        name = el.find('span', class_='pruduct-name-bottom').text
        link = el.find('a').get('href')
        price = el.find('span', class_='regular-price').text
        BASE.append([link, name, price])
    return(BASE)

def pars2():
    for k in range(len(BASE)):
        s1 = BASE[k]
        s2 = s1[0]
        r = requests.get(s2, headers=HEADERS)
        if r.status_code == 200:
            soup = BeautifulSoup(r.text, 'lxml')
            block = soup.find('div', class_='atributes-description')
            articl = block.find('span', class_='sku').text
            packaging = block.find('span', class_= 'type-packaging-name').text
            BASE[k].extend([articl])
            BASE[k].extend([packaging])
    return(BASE)

for i in range(1, 1000):
    if i == 1:
        SITE = SITE
        a = pars1()
    else:
        SITE = f'https://smallpacking.agrosem.ua/shop/page/{i}/'
        try:
            a = pars1()
        except AttributeError:
            break

result = pars2()

import sqlite3

conn = sqlite3.connect('db_agrosem.sqlite3')

curs = conn.cursor()
curs.execute("CREATE TABLE cargo (id_ INT, articl TEXT, name TEXT, price TEXT, packaging TEXT)")

for i in range(len(BASE)):
    a = BASE[i]
    id_ = int(i+1)
    articl = a[3]
    name = a[1]
    price = a[2]
    packaging = a[4]
    conn = sqlite3.connect('db_agrosem.sqlite3')
    curs = conn.cursor()
    s = "INSERT INTO cargo VALUES (?, ?, ?,  ?, ?)"
    curs.execute(s, (id_, articl, name, price, packaging))
    conn.commit()
    conn.close()
```

13. Підрахуйте кількість замовлень для кожного клієнта використовуючи підзапити

```python
import sqlite3 
 
conn = sqlite3.connect('tysql.sqlite')
curs = conn.cursor()

s = '''SELECT Customers.cust_id, Customers.cust_name, count(Orders.cust_id) AS Q 
       FROM Customers LEFT JOIN Orders On Customers.cust_id = Orders.cust_id 
       GROUP BY Orders.cust_id 
       ORDER BY Customers.cust_id'''

curs.execute(s)     
resalt = curs.fetchall() 
print('Таблиця з кількістю замовлень для кожного клієнта:\n')
for i in range(len(resalt)):
    print(f'{resalt[i]}\n')
conn.close()
```

14. Виведіть список клієнтів та їх замовлення (використайте JOIN)

```python
import sqlite3 
 
conn = sqlite3.connect('tysql.sqlite')
curs = conn.cursor()

s = '''SELECT Customers.cust_name, Orders.order_num 
       FROM Customers INNER JOIN Orders ON Customers.cust_id = Orders.cust_id'''

curs.execute(s)     
resalt = curs.fetchall()     
for i in range(len(resalt)):
    print(f'{resalt[i]}\n')
conn.close()
```

15. Зробіть два запити: список товарів, ціна яких менша 5 та список товарів, ціна який більша рівна 5. Використайте UNION для об'єднання цих запитів

```python
import sqlite3 
 
conn = sqlite3.connect('tysql.sqlite')
curs = conn.cursor()

s = '''SELECT Products.prod_name, Products.prod_price 
       FROM Products 
       WHERE Products.prod_price < 5 
       UNION 
       SELECT Products.prod_name, Products.prod_price 
       FROM Products 
       WHERE Products.prod_price >= 5 
       ORDER BY Products.prod_price'''

curs.execute(s)     
resalt = curs.fetchall()     
for i in range(len(resalt)):
    print(f'{resalt[i]}\n')
conn.close()
```
