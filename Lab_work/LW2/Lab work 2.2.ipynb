### Завдання 1

Для трьох введених натуральних чисел (користувач вводить числа через пробіл) знайдіть мінімальне та максимальне

```python
numbers = input('Введіть три натуральних числа через пробіл: ')

while True:
    N = numbers.split()
    if len(N) != 3:
        print ('Числа введено не коректно\n')
        numbers = input('Введіть три натуральних числа через пробіл: ')   
    elif len(N) == 3:
        try:
            n0 = int(N[0])
            n1 = int(N[1])
            n2 = int(N[2])
            break
        except ValueError:
            print ('Числа введено не коректно\n')
            numbers = input('Введіть три натуральних числа через пробіл: ')
            continue
        else:
            break

print('Мінімальне із введених чисел -',min(n0, n1, n2))
print('Максимальне із введених чисел -',max(n0, n1, n2))
```

Якщо введено не коректні дані, виведіть повідомлення про помилку

```python
def error_age():
    
    while True:

        try:
            age = float(input('Введіть вік спортсмена (від 0 до 100): '))

        except ValueError:
            print ('Не коректно введено вік спортсмена.\n')
            continue
        else:
            break
    return age 

sex = input('Введіть стать спортсмена (w або m): ')

while True:
    
    if len(sex)>=2 or (sex[0] != 'w' and sex[0] != 'm'):
        print ('Не коректно введено стать спортсмена.\n')
        sex = input('Введіть стать спортсмена (w або m): ') 
    else:
        break

age = error_age()
        
while True:
    
    if age < 0 or age > 100:
        print ('Не коректно введено вік спортсмена.\n')
        age = error_age()
    else:
        break

if sex[0] == 'w' and age >= 18:
    print('Спотсмен-дівчина може брати учать у змаганні.')

elif sex[0] =='m' and age >= 16:
    print('Спотсмен-хлопець може брати учать у змаганні.')
    
else:
    print('Спотсмен не має права брати учать у змаганні.')
```

```python
import random

#генеруємо 1000 чисел в діапазоні від 0 до 10
rand_num = [random.randint(0,10) for i in range(1000)]
# print(rand_num)

def f_min():
    """Функція обчислює мінімальне значення послідовності"""
    a_min = rand_num[0]
    for i in range(len(rand_num)):
        if int(rand_num[i]) < a_min:
            a_min = rand_num[i]
    return a_min

def f_max():
    """Функція обчислює максимальне значення послідовності"""
    b_max = 0
    for i in range(len(rand_num)):
        if int(rand_num[i]) > b_max:
            b_max = rand_num[i]
    return b_max

def f_s():
    """Функція обчислює середнє значення послідовності"""
    s = 0
    for i in range(len(rand_num)):
        s = s +int(rand_num[i])
    f_s = s / len(rand_num)
    return f_s

def f_m():
    """Функція обчислює медіану послідовності"""
    sort = sorted(rand_num)
    if len(sort) %2 != 0:
        m = sort[len(rand_num)//2]      
    else:
        m = (int(sort[len(rand_num)//2-1]) + int(sort[len(rand_num)//2]))/2
    return m

def f_k():
    """Функція виводить частоту кожної цифри в послідовності"""
    S = {}
    for i in range(len(rand_num)):
        S.setdefault(rand_num[i], rand_num.count(rand_num[i]))
    print(f'Частота кожної цифри в послідовності:') 
    for key in S:
        print(key, S[key])   
    return S
        
def f_sk():
    """Функція обчислює середнє кадратичне відхиленн"""
    s = 0
    for i in range(len(rand_num)):
        s = s +int(rand_num[i])
    f_s = s / len(rand_num)
    s2 = 0
    for i in range(len(rand_num)):
        s2 = s2 + (int(rand_num[i]) - f_s)**2
    f_sk = (s2 / (len(rand_num) - 1))**(1/2)
    return f_sk

a_min = f_min()
print(f'Мінімальне значення послідовності -', a_min)

b_max = f_max()
print(f'Максимальне значення послідовності -', b_max)

f_s = f_s()
print(f'Середнє значення послідовності -', f_s)

m = f_m()
print(f'Медіана послідовності дорівнює -', m)            

f_sk = f_sk()
print(f'Середнє квадратичне відхилення -', f_sk)

k = f_k()
```
