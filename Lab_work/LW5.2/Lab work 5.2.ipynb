**Виведіть зведену таблицю по браузерах(Browser Info), посаді(Job), та кількості транзакцій :**

**Створіть нову колонку "Actual price", яка утворюється із "Purchase Price" та націнки за принципом:**
- якщо "Purchase Price" > 50, націнка 20%
- якщо "Purchase Price" > 100, націнка 10%
- в інших випадках націнка 30%
   

```python
df.loc[df['Purchase Price'] > 100, 'Actual price'] = df.loc[df['Purchase Price'] > 100, 'Purchase Price']*1.1

df.loc[(df['Purchase Price'] > 50) & (df['Purchase Price'] <= 100), 'Actual price'] = \
df.loc[(df['Purchase Price'] > 50) & (df['Purchase Price'] <= 100), 'Purchase Price']*1.2


df.loc[df['Purchase Price'] <= 50, 'Actual price'] = df.loc[df['Purchase Price'] <= 50, 'Purchase Price']*1.3

df[['Purchase Price', 'Actual price']]
```
