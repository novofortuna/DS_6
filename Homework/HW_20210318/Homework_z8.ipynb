### Завдання 8
На карантині я розпочав бігові тренування. В перший день пробіг 500 метрів. Кожний наступний день дистанцію збільшую на 10%. Відомо, що карантин триває 30 днів.  

* Яку дистанцію я пробіжу в останній день карантину?
        
* Скільки днів потрібно бігати, щоб сумарна дистанція за всі дні перевищила довжину марафону (42 км)?

```python
s = [500]         # список, в який бедемо записувати скільки людина пробігла кожного дня із зазначенням інформації за 1-й день
si = s[0]         # відстань, яку людина пробігла в 1-й день

for i in range(29):
    si = (si) * 1.1    
    s.append(si)

print('В останній день карантину людина пробігла ' + str(round((s[29]),2)) + ' метрів')

sz = 0            # в sz будемо рахувати загальну відстань, яку людина пробіжить за i днів 
for i in range(29):
    sz = sz + s[i]
    if sz > 42000:
        print('Потрібно бігати ' + str(i+1) + ' дні, щоб сумарна дистанція перевищила довжину марафону')
        print('За ' + str(i+1) + ' дні людина пробіжить ' +  str(round(sz, 2)) + ' метрів')
        break
```
