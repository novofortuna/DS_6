У файлі мітиться інформація про товарні позиції деякого інтернет-магазину, а саме:
	Артикул - унікальний номер товарної позиції
	Найменування - найменування товару
	Наявність - чи є товарна позиція в даний момент
	Ціна - вартість товарної позиції

Завдання 1

		- кількість товарних позицій у файлі
		- кількість товарних позицій, що в наявності
    	- ціна та назва найдорожчої товарної позиції 

Завдання 2
	Інформацію із файлу запишіть у словник, де ключем буде артикул, а значення для ключа - список, що містить три елементи: Найменування, Наявність, Ціна
	Даний словник серіалізуйте за допомогою бібліотеки:
		- pickle
		- json
	Збережіть ці файли.

Завдання 3
	Відкрийте серіалізований файл (що ви зробили вище). 
	Напишіть програму, що запитує артикул товарної позиції, а повертає стрічку json
	Приклад програми:

		----------------**********----------------
		Введіть артикул запчастини:

		>>>>3963
		{'part' : '3963', 'name' : 'Очиститель двигателя (спрей) Liqui Moly 0.4л',  'status' : 'В наличии', 'price' :'277 грн'}

Завдання 4
	Знайдіть всі товари, у яких аромат яблука, та виведіть їх. 
	(Врахуйте що слово "яблоко" може відмінятися в залежності від речення) 
	

```python
f = open('Price.txt', encoding = 'UTF-8')

f1 = f.readlines()                    # f1 список из строк файла

f2 = f1[1:]                           # f2 список из строк файла без 1-й строки

l = len(f2)
print('У файлі ' + str(l) + ' товарних позицій\n')

max = 0                               # в переменую max будем записывать максимальную цену
name = ()                             # в переменую name будем записывать наименование товара с максимальной ценой   
e = 0                                 # e - счетчик продукции, которая есть в наличии 
n = 0                                 # n - счетчик продукции, которой нет в наличии 
ndict = {}                            # создаем пустой словарь

for i in range(l):
    t1 = f2[i].replace('\n', '')      # сначала удаляем \n (в последней строчке его нет) 
    t2 = t1.replace('грн', '')        # удаляем грн
    t3 = t2.split('|')                # разбиваем строчки по | 
    
    if t3[2] == 'В наличии':
        e = e + 1
    else:
        n = n + 1
    
    m1 = float(t3[3])
    if m1 > max:
        max = m1
        name = t3[1]
    
    ndict[t3[0]] = t3[1:]             # формируем словарь

print('Ціна найдорожчої товарної позиції '+ str(max) + ' грн.\n')
print('Найдорожча товарна позиція - '+ str(name) + '\n')
print('Кількість товарних позицій, що є в наявності - ' + str(e) + '\n')
print('Кількість товарних позицій, яких немає в наявності - ' + str(n) + '\n')

import pickle

ndict

with open('ndict.pickle', 'wb') as f:
    pickle.dump(ndict,f)

import json 

ndict

with open('ndict.json', 'w') as f:     
    json.dump(ndict, f) 
    
f.close()
```

```python
import json    

with open('ndict.json', 'r') as f:
    dict2 = json.load(f)

art = input('Введіть артикул для пошуку інформації \n')

z = dict2.get(art, 'Такий артикул відсутній')

if art in dict2:
    str_json = {
        'part':art,
        'name':z[0],
        'status':z[1],
        'price':z[2]+'грн.'
    }
    print(str_json)
else:
    print(z)
print('\n')

dict3 = {}              # новый словарь в который будем записывать товары в названии которых есть яблоко

for key, val in dict2.items():
    n = dict2.get(key)[0].count('яблок')    
    if n > 0:
        dict3[key] = val

print('Всі товари, в назві яких є "яблуко": ')

for key in dict3:
    print(key, dict3[key])
```
